```python
# reference: http://myinspirationinformation.com/uncategorized/audio-signals-in-python/
def plot_fft(sample, rate):
    sample_fft = fft(sample[:,0])
    n = len(sample_fft)
    sample_fft = sample_fft[0:int(n/2)]  # discard the imaginary part
    freqArray = np.arange(0, (n/2), 1.0) * (rate*1.0/n)  # get the frequencies in Hz
    plt.plot(freqArray/1000, 10*np.log10(np.abs(sample_fft)))
    plt.xlabel('Frequency (kHz)')
    plt.ylabel('Power (dB)')
    plt.show()
    print(np.argmax(np.abs(sample_fft)))
```

```python
plot_fft(arr_g, rate_g)
```
